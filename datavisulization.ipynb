# Network graph

```python
import plotly.graph_objects as go
import networkx as nx
from sklearn.datasets import load_iris


iris = load_iris()

G = nx.Graph()
for i in range(len(iris.data)):
    G.add_node(i, sepal_length=iris.data[i][0], sepal_width=iris.data[i][1])

for i in range(len(iris.data)):
    for j in range(i + 1, len(iris.data)):
        similarity = abs(iris.data[i][0] - iris.data[j][0]) + abs(iris.data[i][1] - iris.data[j][1])
        if similarity < 1.0:
            G.add_edge(i, j)

            
edge_x = []
edge_y = []
for edge in G.edges():
    x0, y0 = G.nodes[edge[0]]['sepal_length'], G.nodes[edge[0]]['sepal_width']
    x1, y1 = G.nodes[edge[1]]['sepal_length'], G.nodes[edge[1]]['sepal_width']
    edge_x.extend([x0, x1, None])
    edge_y.extend([y0, y1, None])

edge_trace = go.Scatter(
    x=edge_x, y=edge_y,
    line=dict(width=0.5, color='#888'),
    hoverinfo='none',
    mode='lines')

node_x = [G.nodes[node]['sepal_length'] for node in G.nodes()]
node_y = [G.nodes[node]['sepal_width'] for node in G.nodes()]

node_trace = go.Scatter(
    x=node_x, y=node_y,
    mode='markers',
    hoverinfo='text',
    marker=dict(
        showscale=True,
        colorscale='YlGnBu',
        reversescale=True,
        color=[],
        size=10,
        colorbar=dict(
            thickness=15,
            title='Node Connections',
            xanchor='left',
            titleside='right'
        ),
        line_width=2))

node_adjacencies = []
node_text = []
for node, adjacencies in enumerate(G.adjacency()):
    node_adjacencies.append(len(adjacencies[1]))
    node_text.append('# of connections: '+str(len(adjacencies[1])))

node_trace.marker.color = node_adjacencies
node_trace.text = node_text

fig = go.Figure(data=[edge_trace, node_trace],
                layout=go.Layout(
                    title='<br>Network graph made with Python',
                    titlefont_size=16,
                    showlegend=False,
                    hovermode='closest',
                    margin=dict(b=20, l=5, r=5, t=40),
                    annotations=[dict(             
                        showarrow=False,
                        xref="paper", yref="paper",
                        x=0.005, y=-0.002)],
                    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False))
                )
fig.show()
```

# Radar Chart

```python
import plotly.express as px
import pandas as pd
df = pd.DataFrame(dict(
    r=[5.1, 3.5, 1.4, 0.2],
    theta=['sepal length','sepal width','petal length', 'petal width']))
fig = px.line_polar(df, r='r', theta='theta', line_close=True)
fig.update_traces(fill='toself')
fig.show()
```

```python
import pandas as pd
import plotly.express as px
from sklearn.datasets import load_iris


iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df['species'] = iris.target_names[iris.target]

iris_long = pd.melt(iris_df, id_vars=['species'], value_vars=iris.feature_names, var_name='feature', value_name='value')

fig = px.line_polar(iris_long, r='value', theta='feature', line_close=True, color='species',
                    title='Radar Chart for Iris Dataset Features',
                    labels={'feature': 'Feature', 'value': 'Value'})

fig.show()
```
